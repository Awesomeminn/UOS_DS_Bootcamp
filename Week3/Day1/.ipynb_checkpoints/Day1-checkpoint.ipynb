## 2. 데이터 핸들링
### 데이터 합치기: combine
- pandas 이용

```python
columns = ['name', 'age', 'gender', 'job']

user1 = pd.DataFrame([['alice', 19, "F", "student"],
                      ['john', 26, "M", "student"]],
                     columns=columns)
user1
```

```python
user2 = pd.DataFrame([['eric', 22, "M", "student"],
                      ['paul', 58, "F", "manager"]],
                     columns=columns)
user2
```

```python
user3 = pd.DataFrame(dict(name=['peter', 'julie'],
                          age=[33, 44],
                          gender=['M', 'F'],
                          job=['engineer', 'scientist']))
user3
```

```python
pd.concat([user1, user2])
```

```python
users = pd.concat([user1, user2, user3])
users
```

- numpy 이용

```python
import numpy as np

arr = np.arange(12).reshape((3, 4))
arr
```

```python
np.vstack([arr,arr])
```

```python
np.hstack([arr,arr])
```

```python
np.concatenate([arr, arr], axis=1)
```

- merge 이용
#### default : inner join
    - inner join     pd.merge(df1, df2)
    - outer join     pd.merge(df1, df2, how = 'outer')
    - left join      pd.merge(df1, df2, how = 'left')
    - right join     pd.merge(df1, df2, how = 'right')

```python
user4 = pd.DataFrame(dict(name=['alice', 'john', 'eric', 'julie'],
                          height=[165, 180, 175, 171]))
user4
```

```python
users2 = pd.merge(users, user4, on="name")
users2
```

```python
users3 = pd.merge(users, user4, on="name", how='outer')
users3
```

```python
pd.merge(users, user4, left_on = 'name', right_on = 'name')
```

### <참고> 표준화와 Min-Max Scaling

```python
# 예제 데이터 불러오기

from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples= 200, centers= 5, random_state=4, cluster_std=1.5)

import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
```

```python
# 표준화 진행

## StandardScaler 이용
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
plt.scatter(X_scaled[:,0],X_scaled[:,1])
plt.title("StandardScaler")
```

```python
# Min-Max Scaling

## MinMaxScaler 이용
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
plt.scatter(X_scaled[:,0],X_scaled[:,1])
plt.title("MinMaxScaler")
```

## 6. 질적 데이터 요약

```python
Score = pd.read_csv("/content/drive/MyDrive/단기기초통계/Day1/data/Score.csv")
Score.describe()
```

### 요인 변수 설정

```python
Score["Dept"] = Score["Dept"].astype("str")
Score["Gender"] = Score["Gender"].astype("str")
Score.describe(include = "all")
```

### 레이블 붙이기

```python
DeptLabel = {"1": "Stat", "2": "Math"}
Score["Dept"] = Score["Dept"].map(DeptLabel)
GenderLabel = {"1": "남자", "2" : "여자"}
Score["Gender"] = Score["Gender"].map(GenderLabel)

print(Score.groupby("Dept").size())
print(Score.groupby("Gender").size())
```

### 빈도표
- crosstab 이용

```python
Drink = pd.read_csv("/content/drive/MyDrive/단기기초통계/Day1/data/Drink.csv", encoding="CP949")

# 1차원 빈도표

pd.crosstab(index = Drink["Age"], columns = "count")
```

```python
# 2차원 분할표

pd.crosstab(index = Drink["Age"], columns = Drink["Drink"])
```

```python
pd.crosstab(index = Drink["Age"], columns = Drink["Drink"], margins = True, margins_name = "합계")
```

```python
# 전체 비율 출력

pd.crosstab(index = Drink["Age"], columns = Drink["Drink"], margins = True, margins_name = "합계", normalize = "all")
```

```python
# 행 비율 출력

pd.crosstab(index = Drink["Age"], columns = Drink["Drink"], normalize = "index")
```

## 7. 그래프를 이용한 질적 데이터 요약
### 막대도표

```python
ChartTable = Drink.groupby("Drink").size()
ChartTable.plot(kind = "bar", x = ChartTable.index, y = ChartTable.values, title = "Drink Preference")
```

```python
import matplotlib.pyplot as plt

plt.bar(ChartTable.index, ChartTable.values)
plt.title("Drink Preference")
```

### 원도표

```python
ChartTable.plot(kind = "pie", x = ChartTable.index, y = ChartTable.values, title = "Drink Preference")
```

```python
plt.pie(ChartTable.values, labels = ChartTable.index, autopct = "%1.1f%%")
plt.title("Drink Preference")
```

### 파레토 도표

```python
from matplotlib.ticker import PercentFormatter

df = Drink.Drink.value_counts()
df_cumpct = df.cumsum() / df.sum()*100

fig, ax = plt.subplots()
ax.bar(df.index, df.values, color = "C0")
ax2 = ax.twinx()
ax2.plot(df.index, df_cumpct, color = "C1", marker = "D", ms = 7)
ax2.yaxis.set_major_formatter(PercentFormatter())
ax.tick_params(axis = "y", colors = "C0")
ax2.tick_params(axis = "y", colors = "C1")
ax.set_title("Drink Preference")
```
