```python
x = 1000
print("초깃값: ", x)
x += 2
print("x += 2 후: ", x)
x -= 2
print("x -= 2 후: ", x)
x *= 2
print("x *= 2 후: ", x)
x /= 2
print("x /= 2 후: ", int(x))
```
